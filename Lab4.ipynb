```python
import numpy as np
import matplotlib.pyplot as plt

# Function to minimize: f(x) = x^2
def f(x):
    return x**2

# Derivative of the function: f'(x) = 2x
def f_prime(x):
    return 2*x

# Gradient Descent function
def gradient_descent(starting_point, learning_rate, iterations):
    x = starting_point
    path = [x]  # To store the points for visualization

    for i in range(iterations):
        gradient = f_prime(x)
        x = x - learning_rate * gradient
        path.append(x)

    return path

# Parameters for Gradient Descent
starting_point = 10  # Starting at x = 10
learning_rate = 0.1  # Learning rate (step size)
iterations = 25      # Number of iterations
# Running gradient descent
path = gradient_descent(starting_point, learning_rate, iterations)
# Visualization
x_vals = np.linspace(-10, 10, 400)
y_vals = f(x_vals)

plt.figure(figsize=(8, 6))
plt.plot(x_vals, y_vals, label="f(x) = x^2")
plt.scatter(path, f(np.array(path)), color='red', label="Gradient Descent steps", zorder=5)
plt.plot(path, f(np.array(path)), color='red', linestyle='--', alpha=0.6)
plt.title("Gradient Descent Visualization")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Task 4 Experiment with Different Activation Functions:
import numpy as np
import matplotlib.pyplot as plt

# Function to minimize: f(x) = x^2
def f(x):
    return x**2

# Derivative of the function: f'(x) = 2x
def f_prime(x):
    return 2*x

# Gradient Descent function
def gradient_descent(starting_point, learning_rate, iterations):
    x = starting_point
    path = [x]  # To store the points for visualization

    for i in range(iterations):
        gradient = f_prime(x)
        x = x - learning_rate * gradient
        path.append(x)

    return path

# Parameters for Gradient Descent
starting_point = 10  # Starting at x = 10
learning_rate = 0.1  # Learning rate (step size)
iterations = 25      # Number of iterations
# Running gradient descent
path = gradient_descent(starting_point, learning_rate, iterations)
# Visualization
x_vals = np.linspace(-10, 10, 400)
y_vals = f(x_vals)

plt.figure(figsize=(8, 6))
plt.plot(x_vals, y_vals, label="f(x) = x^2")
plt.scatter(path, f(np.array(path)), color='red', label="Gradient Descent steps", zorder=5)
plt.plot(path, f(np.array(path)), color='red', linestyle='--', alpha=0.6)
plt.title("Gradient Descent Visualization")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show()

# XOR dataset
X = np.array([[0, 0],
              [0, 1],
              [1, 0],
              [1, 1]])
y = np.array([[0], [1], [1], [0]])

# Sigmoid and its derivative
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_deriv(x):
    s = sigmoid(x)
    return s * (1 - s)

# ReLU and its derivative
def relu(x):
    return np.maximum(0, x)

def relu_deriv(x):
    return (x > 0).astype(float)

def train_nn(X, y, activation='sigmoid', epochs=10000, lr=0.1):
    np.random.seed(0)
    input_size = 2
    hidden_size = 2
    output_size = 1

    # Initialize weights
    W1 = np.random.randn(input_size, hidden_size)
    b1 = np.zeros((1, hidden_size))
    W2 = np.random.randn(hidden_size, output_size)
    b2 = np.zeros((1, output_size))

    losses = []

    # Select activation functions
    if activation == 'sigmoid':
        act = sigmoid
        act_deriv = sigmoid_deriv
    elif activation == 'relu':
        act = relu
        act_deriv = relu_deriv
    else:
        raise ValueError("Unsupported activation")

    for epoch in range(epochs):
        # Forward pass
        z1 = np.dot(X, W1) + b1
        a1 = act(z1)
        z2 = np.dot(a1, W2) + b2
        a2 = sigmoid(z2)  # Output layer uses sigmoid

        # Loss (binary cross-entropy)
        loss = np.mean((a2 - y)**2)
        losses.append(loss)

        # Backpropagation
        dz2 = (a2 - y) * sigmoid_deriv(z2)
        dW2 = np.dot(a1.T, dz2)
        db2 = np.sum(dz2, axis=0, keepdims=True)

        da1 = np.dot(dz2, W2.T)
        dz1 = da1 * act_deriv(z1)
        dW1 = np.dot(X.T, dz1)
        db1 = np.sum(dz1, axis=0, keepdims=True)

        # Update weights
        W1 -= lr * dW1
        b1 -= lr * db1
        W2 -= lr * dW2
        b2 -= lr * db2

    # Final predictions
    predictions = (a2 > 0.5).astype(int)
    accuracy = np.mean(predictions == y)

    return losses, accuracy

# Train with Sigmoid activation
losses_sigmoid, acc_sigmoid = train_nn(X, y, activation='sigmoid')
print("Sigmoid Accuracy:", acc_sigmoid)

# Train with ReLU activation
losses_relu, acc_relu = train_nn(X, y, activation='relu')
print("ReLU Accuracy:", acc_relu)

# Plot loss curves
plt.plot(losses_sigmoid, label='Sigmoid')
plt.plot(losses_relu, label='ReLU')
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Loss Curve Comparison")
plt.legend()
plt.grid(True)
plt.show()
```
